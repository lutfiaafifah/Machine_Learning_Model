### Identify Outlier
* High outlier ditemukan di hampir semua kolom kecuali kolom 'alcohol'
* Low outlier ditemukan di beberapa kolom, yaitu 'fixed acidity', 'citrid acid', 'chlorides', 'total sulfur dioxide', 'pH', dan 'quality'
* Pada kolom 'alcohol' tidak ditemukan adanya outlier, baik itu low outlier maupun high outlier

```python
for i in range(len(df_wine.columns)):
    if (df_wine[df_wine.columns[i]].dtypes in ['int64','float64']):
        print('\nAttribute-',df_wine.columns[i],':',df_wine[df_wine.columns[i]].dtypes)
        Q1=df_wine[df_wine.columns[i]].quantile(0.25)
        print('Q1',Q1)
        Q3=df_wine[df_wine.columns[i]].quantile(0.75)
        print('Q3',Q3)
        IQR=Q3-Q1
        print('IQR',IQR)
        min=df_wine[df_wine.columns[i]].min()
        max=df_wine[df_wine.columns[i]].max()
        min_IQR=Q1-1.5*IQR
        max_IQR=Q3+1.5*IQR
        if (min<min_IQR):
            print('Low outlier is found < ',min_IQR,' - ', min)
        if (max>max_IQR):
            print('High outlier is found > ',max_IQR,' - ', max)
```

### Drop outliers which have big gap with others

### Binning process
* proses binning dilakukan untuk membagi label di kolom 'quality' menjadi dua kategori yaitu low dan high
* low quality yaitu terdiri dari label 1 sampai 5 dan direpresentasikan dengan angka 0, sedangkan high quality terdiri dari label 6 sampai 10 dan direpresentasikan dengan angka 1

```python
bins = [1, 6, 10]
quality_cat=[0, 1]
df_wine['quality_category'] = pd.cut(df_wine['quality'], bins, right=False, labels=quality_cat)
df_wine.sample(10)
```

### 4. Modeling

### Membagi dataset menjadi data training dan data testing

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

source = df_wine.iloc[:,:-1] #all columns except the last one
target = df_wine['quality_category'] #only the last column

sv_train, sv_test, tv_train, tv_test = train_test_split(source, target, random_state=3, test_size=0.2)
print("\nData Training:\n")
print(sv_train.head())
print(sv_train.shape)

print("\nData Testing:\n")
print(sv_test.head())
print(sv_test.shape)
```

### Evaluasi score tanpa meng-handle imbalanced data

#### KNN Classifier (k=50)

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

knn_model = KNeighborsClassifier(n_neighbors=50)

knn_model.fit(sv_train, tv_train)
pred = knn_model.predict(sv_test)
print("[INFO] evaluating...")
print(classification_report(tv_test, pred))
```

#### KNN Classifier (k=20)

```python
knn_model = KNeighborsClassifier(n_neighbors=20)

knn_model.fit(sv_train, tv_train)
pred = knn_model.predict(sv_test)
print("[INFO] evaluating...")
print(classification_report(tv_test, pred))
```

#### KNN Classifier (k=10)

```python
knn_model = KNeighborsClassifier(n_neighbors=10)

knn_model.fit(sv_train, tv_train)
pred = knn_model.predict(sv_test)
print("[INFO] evaluating...")
print(classification_report(tv_test, pred))
```

#### KNN Classifier (k=5)

```python
knn_model = KNeighborsClassifier(n_neighbors=5)

knn_model.fit(sv_train, tv_train)
pred = knn_model.predict(sv_test)
print("[INFO] evaluating...")
print(classification_report(tv_test, pred))
```

#### KNN Classifier (k=3)

```python
knn_model = KNeighborsClassifier(n_neighbors=3)

knn_model.fit(sv_train, tv_train)
pred = knn_model.predict(sv_test)
print("[INFO] evaluating...")
print(classification_report(tv_test, pred))
```

#### Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

nb_model = GaussianNB()

nb_model.fit(sv_train, tv_train)
pred = nb_model.predict(sv_test)
print("[INFO] evaluating...")
print(classification_report(tv_test, pred))
```

### Handling imbalanced data with oversampling method - SMOTE
